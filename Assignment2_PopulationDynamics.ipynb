```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def popdyn(dt, a, b, c, d, t_max, x_0, y_0):
    
    #x - population of prey
    #y - population of predator
    #x_0 - initial population of prey
    #y_0 - initial populatio of predator
    
    #dt - time step
    #t_max - stop time of simulation
    
    #a - reproduction rate of prey
    #b - death rate of prey (getting eaten by eating predator)
    #c - death rate of predator (no prey to eat)
    #d - reproduction rate of predator (eating a prey)


    x = []
    y = []
    time = []
    
    time = np.arange(0,t_max+dt,dt) #array of time steps
    
    #initialising x and y
    x = [None]*len(time)
    y = [None]*len(time)
    
    x[0] = 3.0
    y[0] = 5.0
    
    #difference equation
    for t in range(0,len(time)-1):
        x[t+1] = x[t] + (dt * ((a*x[t]) - (b*x[t]*y[t])))
        y[t+1] = y[t] + (dt * ((-c*y[t]) + (d*x[t]*y[t])))
        
    return x,y,dt,time,a
```

# Q1, Q2, Q3, Q5: a = 1

```python
#Q1,Q2,Q3,Q5

x,y,dt,time,a = popdyn(dt = 0.001, a = 1.0, b = 0.5, c = 2.0, d = 1.0, t_max = 20, x_0 = 3.0, y_0 = 5.0)
print ("x(12) = ", x[int(12/dt)])
print ("y(10) = ", y[int(10/dt)])
print('max (x, y) = (',max(x),',',max(y),')')
print('min (x, y) = (',min(x),',',min(y),')')

x,y,dt,time,a = popdyn(dt = 0.001, a = 1.0, b = 0.5, c = 2.0, d = 1.0, t_max = 10, x_0 = 3.0, y_0 = 5.0)
fig, ax = plt.subplots()
ax.plot(time,x,label='x')
ax.plot(time,y,label='y')
ax.set_xlabel('time')
ax.set_ylabel('population')
ax.legend(loc ='upper left')
```
